# Pivot Table 

```python
dr = sns.load_dataset("tips")
dr.head()
```

```python
dr.groupby("sex")[["total_bill"]].mean()
```

```python
dr.groupby(["sex","smoker"])[["total_bill"]].mean().unstack()
```

```python
dr.pivot_table(index = "sex" , columns =  "smoker" , values = "total_bill")
```

```python
# aggfunc
```

```python
dr.pivot_table(index = "sex" , columns =  "smoker" , values = "total_bill" , aggfunc = "sum")
```

```python
dr.pivot_table(index = "sex" , columns =  "smoker" , values = "total_bill" , aggfunc = "std")
```

```python
# use all col withot value fun use
dr.pivot_table(index = "sex" , columns =  "smoker")
```

```python
# multidimensional 
dr.head()
```

```python
dr.pivot_table(index = ["sex" , "smoker"] ,columns =  ["day" , "time"])
```

```python
v = pd.read_csv("titanic.csv")
v["Name"]
```

```python
# Common Functions
# lower/upper/capitalize/title


l = v["Name"].str.lower()
u = v["Name"].str.upper()
c = v["Name"].str.capitalize()
t = v["Name"].str.title()
```

```python
# len

v["Name"].str.len()
```

```python
# strip use for remove space

v["Name"].str.strip()

```

```python
v["FN"]=v["Name"].str.split(",")
v.head()
```

```python
# split -> get
v["Name"].str.split(",").str.get(0)
```

```python
v[["title" , "FN"]] = v["Name"].str.split(",").str.get(1).str.strip().str.split(" ", n=1 , expand = True)
v["title"].value_counts()
```

```python
# Replace
```

```python
v['title'] = v['title'].str.replace('Ms.','Miss.',regex=True)
v['title'] = v['title'].str.replace('Mlle.','Miss.', regex=True)
```

```python
v["title"].value_counts()
```

```python
# filtering
# startswith/endswith
```

```python
v[v['FN'].str.endswith('A')]
```
