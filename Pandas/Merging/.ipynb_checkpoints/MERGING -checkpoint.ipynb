```python
Y = pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})

students = pd.concat([students,Y],ignore_index=True)
students.tail()
```

```python
students.merge(x,how = "right" , on = "student_id")
```

```python
# outer join
```

```python

students.merge(x,how = "outer" , on = "student_id").tail(10)
```

```python
total = x.merge(courses,how = "inner" , on = "course_id")["price"].sum()
total
```

```python
# 2. find month by month revenue

month_nov = nov.merge(courses , how = "inner" , on = "course_id")["price"].sum()
month_nov
```

```python
month_nov = dec.merge(courses , how = "inner" , on = "course_id")["price"].sum()
month_nov
```

```python
# method 2
month = pd.concat([nov,dec],keys=['Nov','Dec']).reset_index()
month.merge(courses,on='course_id').groupby('level_0')['price'].sum()
```
