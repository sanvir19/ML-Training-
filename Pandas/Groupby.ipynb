```python
import numpy as np
import pandas as pd
```

```python
movie = pd.read_csv("imdb-top-1000.csv")
movie.head(5)
```

```python
genre = movie.groupby("Genre")
```

```python
genre.sum()
```

```python
# find the top 3 genres by total earning

movie.groupby("Genre").sum()["Gross"].sort_values(ascending  = False).head(3)
```

```python
# find the genre with highest avg IMDB rating

movie.groupby("Genre").mean()["IMDB_Rating"].sort_values(ascending = False).head()
```

```python
movie.value_counts("Genre")
```

```python
# find director with most popularity
movie.groupby("Director").sum()["No_of_Votes"].sort_values(ascending  = False).head()
```

```python
# find the highest rated movie of each genre

movie.groupby("Genre")["IMDB_Rating","Series_Title"].max()
```

```python
# find number of movies done by each actor

movie.groupby("Star1")["Series_Title"].count().sort_values(ascending=False)
```

```python
ipl = pd.read_csv("deliveries.csv")
ipl.head()
```

```python
# find the top 10 batsman in terms of runs

ipl.groupby("batsman").sum()["batsman_runs"].sort_values(ascending = False).head(10)
```

```python
# find the batsman with max no of sixes
```

```python
six = ipl[ipl["batsman_runs"] == 6]
six.groupby("batsman")["batsman_runs"].count().sort_values(ascending = False).head(1)
```

```python
# find batsman with most number of 4's and 6's in last 5 overs
```

```python
x = ipl[(ipl["over"] >= 15) & (ipl["batsman_runs"] >= 4)]

x.groupby("batsman")["batsman_runs"].count().sort_values(ascending = False).head()
```
