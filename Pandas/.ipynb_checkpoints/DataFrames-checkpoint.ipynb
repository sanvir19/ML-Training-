# Filtering a DataFrams

```python
ipl.head(5)
```

```python
# Find all the Final Winners

match =ipl[ ipl["MatchNumber"] == "Final"]
match[["Season" , "WinningTeam"]]


```

```python
ipl[ipl["MatchNumber"] == "Final"][["Season" , "WinningTeam"]]
```

```python
# how many super over finishes have occurred 
ipl[ipl[['SuperOver']] == "Y"].shape
```

```python
# How many match has csk won in kolkata

kolkata = ipl[ipl["City"] == "Kolkata"]
kolkata[kolkata["WinningTeam"] == "Chennai Super Kings"].shape[0]
```

```python
ipl[(ipl["City"] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")]
```

```python
# Toss Winner is match Winner in Percentage

ipl[ipl["TossWinner"] == ipl["WinningTeam"]]
```
